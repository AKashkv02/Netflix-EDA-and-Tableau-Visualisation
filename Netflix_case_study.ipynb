##Analysis of actors

```python
df[['cast']]    #unnesting
```

```python
cast=df['cast'].apply(lambda x: x.split(',')).to_list()
cast_df=pd.DataFrame(cast,index=df['title'])
cast_df
```

```python
cast_df=cast_df.stack()
cast_df
```

```python
cast_df=cast_df.reset_index()
cast_df
```

```python
cast_df.drop(columns='level_1',inplace=True)
cast_df.columns=['title','cast']
cast_df.head()
```

```python
df_final=df.drop(columns=['cast']).merge(cast_df,on='title',how='left')
df_final
```

```python
df_final['cast'].value_counts()[1:11]
```

```python
sns.countplot(y='cast',data=df_final,order=df_final['cast'].value_counts().index[1:11])
```

##Summary
1.Most of the top 10 popluar cast are from India

2.Anupam is casted in 39 movies 

3.Anumpam is more popualar than shah rukh khan ,surprising!




##Genre analysis

```python
df.head()
```

```python
x=df['listed_in'].apply(lambda x: x.split(',')).to_list()
listed_df=pd.DataFrame(x,index=df['title'])
listed_df
```

```python
listed_df=listed_df.stack()
listed_df
```

```python
listed_df=listed_df.reset_index()
```

```python
listed_df
```

```python
listed_df.drop(columns='level_1',inplace=True)
listed_df.columns=['title','genre']
listed_df.head()
```

```python
listed_final=df.merge(listed_df,on='title',how='left')
listed_final
```

```python
listed_final['genre'].value_counts()[0:20]
```

```python
plt.figure(figsize=(16,12))
sns.countplot(y='genre',data=listed_final,order=listed_final['genre'].value_counts().index[0:15])
plt.title('Popular category')
```

1.Most popular categories are international movies,dramas,comedies 

2.if movie or a tv show is of the above category then it is likely to be more popular when compared to other categories
